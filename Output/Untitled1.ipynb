```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://datosmacro.expansion.com/demografia/poblacion/espana-comunidades-autonomas/madrid'
res = requests.get('https://datosmacro.expansion.com/demografia/poblacion/espana-comunidades-autonomas/madrid')
html = res.text
poblacion = BeautifulSoup(html, 'html.parser')
poblacion
```

```python
 poa = poblacion.find_all('tr')
```

```python
poa
```

```python
type(poa)
```

```python
def buscarporfilas(t):
    pob = t.find_all('td')
    try :
        return {
            "Fecha":pob[0].text,
            "Poblacion":pob[4].text
        }
    except:
        return None
```

```python
df =pd.DataFrame(list(filter( lambda x: x ,map(lambda t: buscarporfilas(t), poa))))
df
```

```python
df =df.loc[::-1]
```

```python

df = df[(df.index <= 43) & 
                (df.index>= 6)]
df
```
