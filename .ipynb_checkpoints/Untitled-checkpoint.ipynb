```python
import pandas as pd

df = pd.read_csv("weather_madrid_1997_2015.csv")
df
```

```python
def quitarcolumnas(*x):
    
    return df.drop(list(x),axis = 1)

df =quitarcolumnas(' Max Gust SpeedKm/h'," CloudCover",'Precipitationmm'," Mean Sea Level PressurehPa",'WindDirDegrees', ' Max Sea Level PressurehPa',' Min Sea Level PressurehPa','Dew PointC', 'MeanDew PointC', 'Min DewpointC',' Events')
```

```python
df.columns
```

```python
def renamecolumnas(x):
    
    return df.rename(columns=x)
df =renamecolumnas({'Max TemperatureC':'Max Temp','Mean TemperatureC': 'Mean Temp','Min TemperatureC':'Min Temp','Max Humidity': 'Max Hum',' Mean Humidity':'Mean Hum',' Min Humidity':'Min Hum',' Max VisibilityKm': 'Max Vis Km',' Mean VisibilityKm':'Mean Vis Km',' Min VisibilitykM': 'Min Vis Km',' Max Wind SpeedKm/h':'Max Wind Km/h',' Mean Wind SpeedKm/h': 'Mean Wind Kmh'})
```

```python
df["CET"]=pd.to_datetime(df["CET"])
```

```python
df['day'] = df['CET'].dt.day
df['month'] = df['CET'].dt.month
df['year'] = df['CET'].dt.year
```

```python
print(df.isna().sum())
df.columns
```

```python
df =quitarcolumnas("CET")

```

```python
lista =['day', 'month', 'year','Max Temp', 'Mean Temp', 'Min Temp', 'Max Hum', 'Mean Hum','Min Hum', 'Max Vis Km', 'Mean Vis Km', 'Min Vis Km', 'Max Wind Km/h','Mean Wind Kmh']
df =df[lista]
df
```

```python
#quitarme los 29 de febrero

dropcols = df[(df['day']== 29) & (df['month']==2)]
dropcols.index
df =df.drop(dropcols.index)
df
```

```python
#sustituir si es < que 6 por 1 y si es mayor que 6 > por 2
# bins y concatenate
df.dtypes

```

```python
cutoffs = [0,6,12]
mpg_labels = ['1', '2']
bins = pd.cut(df['month'],cutoffs, labels=mpg_labels)
df["semester"] = bins
df
```

```python
def groupBy(df,columns):
    return (df.groupby(columns).sum())

df =groupBy(df,["year","semester"])
    
    
```

```python
def resetIndex(df):
    df.reset_index(inplace = True)

resetIndex(df)
```
